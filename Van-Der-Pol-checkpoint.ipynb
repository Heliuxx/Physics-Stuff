```python
"""
Explaination:

Vanderpol Oscillation using a While-Loop

@author: Steven Bates
"""

import numpy as np
import matplotlib.pyplot as plt
from collections import deque

mu = 1               #Change this value

x0= 1                          
v0= 0                          
a0= 0
t= 0
dt= 0.0001



x_values= []
v_values= []
a_values= []
t_values= []

while t<= 50:
    a= -x0 -(mu) * (x0 ** 2 - 1) * v0
    v= v0 + a*dt
    x= x0 + v*dt
    x_values.append(x)
    v_values.append(v)
    a_values.append(a)
    t_values.append(t)
    a0= a
    v0= v
    x0= x
    t= t + dt



plt.figure(figsize= (12,8), facecolor= '#fffec4')
plt.plot(x_values, v_values, 'k-', label= '', linewidth= 2) 
#plt.plot(t_values, v_values, 'b-', label= 'Velocity [m/s]', linewidth= 2)
plt.title("Van Der Pol Oscillation (Mu= -5)", color='k', size=18)
plt.xlabel('Position [m]', color= 'k', size=14)
plt.ylabel('Velocity [m/s]', color= 'k', size=14)
plt.legend(loc='best')
plt.grid(color= 'k', linestyle= '--', linewidth=0.6)
plt.show()

plt.savefig('van-der-pol oscillation (mu= -5).png')

```

```python
"""
Created on Mon Jun 11 15:01:24 2018

@author: Steven Bates
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint


mu = 1
t = np.arange(0, 50, 0.0001)
x0=1
v0=0



def vdp(x, time):
    x0 = x[1]
    a = -x[0] -(mu) * (x[0] ** 2 - 1) * x[1]
    vdp_sol = np.array([x0, a])
    return vdp_sol

X= [x0, v0]
Solve = odeint(vdp, X, t)

x = Solve[:,0]  
v = Solve[:,1] 



plt.figure(figsize= (12,8), facecolor= '#fffec4')
plt.plot(x, v, 'k-', label= 'Position [m]', linewidth= 2) 
#plt.plot(t, v, 'b-', label= 'Velocity [m/s]', linewidth= 2)
plt.title("Van Der Pol Oscillation", color='k', size=18)
plt.xlabel('Time [s]', color= 'k', size=14)
plt.ylabel('Position [m], Velocity [m/s]', color= 'k', size=14)
plt.legend(loc='best')
plt.grid(color= 'k', linestyle= '--', linewidth=0.6)
plt.show()





plt.savefig('van-der-pol oscillation.png')

```
